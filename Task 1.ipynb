Question 2 - Do you think we need a machine learning model to differentiate between static activities (laying, sitting, standing) and dynamic activities(walking, walking_downstairs, walking_upstairs)? Look at the linear acceleration 
$ (\text{accx}^2 + \text{accy}^2 + \text{accz}^2) $
for each activity and justify your answer. [0.5 marks]

```python
import numpy as np

def linear_acceleration(accx, accy, accz):
    return (accx**2 + accy**2 + accz**2)

laying = r'C:\Users\Vansh\Desktop\Coding\ML\Assignment 1\Task 1\Combined\Test\LAYING\Subject_2.csv'
sitting = r'C:\Users\Vansh\Desktop\Coding\ML\Assignment 1\Task 1\Combined\Test\SITTING\Subject_4.csv'
standing = r'C:\Users\Vansh\Desktop\Coding\ML\Assignment 1\Task 1\Combined\Test\STANDING\Subject_9.csv'
walking = r'C:\Users\Vansh\Desktop\Coding\ML\Assignment 1\Task 1\Combined\Test\Walking\Subject_10.csv'
walking_down = r'C:\Users\Vansh\Desktop\Coding\ML\Assignment 1\Task 1\Combined\Test\WALKING_DOWNSTAIRS\Subject_12.csv'
walking_up = r'C:\Users\Vansh\Desktop\Coding\ML\Assignment 1\Task 1\Combined\Test\WALKING_UPSTAIRS\Subject_13.csv'

laying_data = np.loadtxt(laying, delimiter=',', skiprows=1, max_rows=500)
sitting_data = np.loadtxt(sitting, delimiter=',', skiprows=1, max_rows=500)
standing_data = np.loadtxt(standing, delimiter=',', skiprows=1, max_rows=500)
walking_data = np.loadtxt(walking, delimiter=',', skiprows=1, max_rows=500)
walking_down_data = np.loadtxt(walking_down, delimiter=',', skiprows=1, max_rows=500)
walking_up_data = np.loadtxt(walking_up, delimiter=',', skiprows=1, max_rows=500)

laying_lin_acc = linear_acceleration(laying_data[:, 0], laying_data[:, 1], laying_data[:, 2])
sitting_lin_acc = linear_acceleration(sitting_data[:, 0], sitting_data[:, 1], sitting_data[:, 2])
standing_lin_acc = linear_acceleration(standing_data[:, 0], standing_data[:, 1], standing_data[:, 2])
walking_lin_acc = linear_acceleration(walking_data[:, 0], walking_data[:, 1], walking_data[:, 2])
walking_down_lin_acc = linear_acceleration(walking_down_data[:, 0], walking_down_data[:, 1], walking_down_data[:, 2])
walking_up_lin_acc = linear_acceleration(walking_up_data[:, 0], walking_up_data[:, 1], walking_up_data[:, 2])

avg_laying = np.mean(laying_lin_acc)
avg_sitting = np.mean(sitting_lin_acc)
avg_standing = np.mean(standing_lin_acc)
avg_walking = np.mean(walking_lin_acc)
avg_walking_down = np.mean(walking_down_lin_acc)
avg_walking_up = np.mean(walking_up_lin_acc)

print("Average Linear Acceleration:")
print(f"Laying: {avg_laying}")
print(f"Sitting: {avg_sitting}")
print(f"Standing: {avg_standing}")
print(f"Walking: {avg_walking}")
print(f"Walking Downstairs: {avg_walking_down}")
print(f"Walking Upstairs: {avg_walking_up}")
```

Observation: 

The linear acceleration values for static activities (Laying, Sitting, Standing) are relatively close to each other (ranging from 1.011 to 1.056).
These values are low compared to dynamic activities, shownig less movement.

The linear acceleration values for dynamic activities (Walking, Walking Downstairs, Walking Upstairs) are higher (ranging from 1.139 to 1.179).
This shows more significant changes in acceleration due to the movements.

The difference in acceleration values between static and dynamic activities is there but not major. 

Human motion can be variable. For example, acceleration while walking can vary based on walking speed. A machine learning model can capture these variations and improve accuracy.

3.1 - Use PCA (Principal Component Analysis) on Total Acceleration 
$ (\text{accx}^2 + \text{accy}^2 + \text{accz}^2) $
 to compress the acceleration timeseries into two features and plot a scatter plot to visualize different class of activities.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
import seaborn as sns
import tsfel

laying = r'C:\Users\Vansh\Desktop\Coding\ML\Assignment 1\Task 1\Combined\Test\LAYING\Subject_2.csv'
sitting = r'C:\Users\Vansh\Desktop\Coding\ML\Assignment 1\Task 1\Combined\Test\SITTING\Subject_4.csv'
standing = r'C:\Users\Vansh\Desktop\Coding\ML\Assignment 1\Task 1\Combined\Test\STANDING\Subject_9.csv'
walking = r'C:\Users\Vansh\Desktop\Coding\ML\Assignment 1\Task 1\Combined\Test\Walking\Subject_10.csv'
walking_down = r'C:\Users\Vansh\Desktop\Coding\ML\Assignment 1\Task 1\Combined\Test\WALKING_DOWNSTAIRS\Subject_12.csv'
walking_up = r'C:\Users\Vansh\Desktop\Coding\ML\Assignment 1\Task 1\Combined\Test\WALKING_UPSTAIRS\Subject_20.csv'

laying_df = pd.read_csv(laying)
sitting_df = pd.read_csv(sitting)
standing_df = pd.read_csv(standing)
walking_df = pd.read_csv(walking)
walking_down_df = pd.read_csv(walking_down)
walking_up_df = pd.read_csv(walking_up)

def total_acc(df):
    df['acc_total'] = (df['accx']**2 + df['accy']**2 + df['accz']**2)
    return df

laying_df = total_acc(laying_df)
sitting_df = total_acc(sitting_df)
standing_df = total_acc(standing_df)
walking_df = total_acc(walking_df)
walking_down_df = total_acc(walking_down_df)
walking_up_df = total_acc(walking_up_df)

laying_df['Label'] = 6
sitting_df['Label'] = 4
standing_df['Label'] = 5
walking_df['Label'] = 1
walking_down_df['Label'] = 3
walking_up_df['Label'] = 2 

combined_df = pd.concat([laying_df, sitting_df, standing_df, walking_df, walking_down_df, walking_up_df], ignore_index=True)

pca_dataframe = pd.DataFrame()
classes = {"WALKING": 1, "WALKING_UPSTAIRS": 2, "WALKING_DOWNSTAIRS": 3, "SITTING": 4, "STANDING": 5, "LAYING": 6}

for label in tuple(classes.values()):
    subset = combined_df[combined_df["Label"] == label]["acc_total"].values
    subset = subset[:len(subset) - (len(subset) % 500)]
    subset = subset.reshape(-1, 500)
    pca = PCA(n_components=2)
    pca_result = pca.fit_transform(subset)
    pca_df = pd.DataFrame(data=pca_result, columns=["PCA1", "PCA2"])
    pca_df["Label"] = label
    pca_dataframe = pd.concat([pca_dataframe, pca_df], ignore_index=True)

label_mapping = {1: "WALKING", 2: "WALKING_UPSTAIRS", 3: "WALKING_DOWNSTAIRS", 4: "SITTING", 5: "STANDING", 6: "LAYING"}
pca_dataframe["Activity"] = pca_dataframe["Label"].map(label_mapping)

sns.set(style="whitegrid")
plt.figure(figsize=(19, 9))
sns.scatterplot(x="PCA1", y="PCA2", hue="Activity", data=pca_dataframe, s=300, palette="tab10")
plt.title("PCA Scatter Plot for Total Acceleration", fontweight='bold', fontsize=20)
plt.xlabel("PCA 1", fontweight='bold', fontsize=20)
plt.ylabel("PCA 2", fontweight='bold', fontsize=20)
plt.legend(title="Activities")
plt.show()
```

3.2 - Next, use TSFEL (a featurizer library) to create features (your choice which ones you feel are useful) and then perform PCA to obtain two features. Plot a scatter plot to visualize different class of activities.

```python
X = combined_df[['accx', 'accy', 'accz']]
y = combined_df['Label']

cfg_file = tsfel.get_features_by_domain()

X_tsfel = tsfel.time_series_features_extractor(cfg_file, X, window_size=500, verbose=0)

num_windows = X_tsfel.shape[0]
X_tsfel["Label"] = y.iloc[::500][:num_windows].values

pca = PCA(n_components=2)
pca_tsfel = pca.fit_transform(X_tsfel.iloc[:, :-1])
pca_tsfel_df = pd.DataFrame(data=pca_tsfel, columns=["PCA_1", "PCA_2"])
pca_tsfel_df["Label"] = X_tsfel["Label"]

label_map = {1: "WALKING", 2: "WALKING_UPSTAIRS", 3: "WALKING_DOWNSTAIRS", 4: "SITTING", 5: "STANDING", 6: "LAYING"}
pca_tsfel_df["Activity"] = pca_tsfel_df["Label"].map(label_map)

sns.set(style="whitegrid")
plt.figure(figsize=(19, 9))
sns.scatterplot(x="PCA_1", y="PCA_2", hue="Activity", data=pca_tsfel_df, s=300, palette="tab10")
plt.title("PCA Scatter Plot for TSFEL Features", fontweight='bold', fontsize=20)
plt.xlabel("PCA 1", fontweight='bold', fontsize=20)
plt.ylabel("PCA 2", fontweight='bold', fontsize=20)
plt.legend(title="Activities")
plt.show()
```
